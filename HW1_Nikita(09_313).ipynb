```python
total_legs = 64

for g in range(0, total_legs // 2 + 1):
    r = (total_legs - 2 * g) // 4
    if 2 * g + 4 * r == total_legs and r >= 0:
        print(f"Гусей: {g}, Кроликов: {r}")
```

```python
from math import factorial

epsilon = 1e-3
n = 0
term = 10**n / factorial(n)
sum_series = term
while abs(term) > epsilon:
    n += 1
    term = 10**n / factorial(n)
    sum_series += term

print("Сумма ряда с заданной точностью:", sum_series)
print("Количество учтённых членов ряда:", n)
```

```python
N = int(input("Введите N (>1): "))
numbers = list(map(int, input(f"Введите {N} целых чисел(ла): ").split()))
smaller_than_left = []
for i in range(1, N):
    if numbers[i] < numbers[i - 1]:
        smaller_than_left.append(numbers[i])

K = len(smaller_than_left)

print("Элементы, меньшие своего левого соседа:", smaller_than_left)
print("Количество таких элементов K:", K)
```

# **Процедуры и функции: группа Proc**

image.png

```python
def sum_odd_digit_3digits():
    total_sum = 0
    for num in range(100, 1000):
        digits = str(num)
        if all(int(d) % 2 == 1 for d in digits):
            total_sum += num
    return total_sum

def count_even_digits(n):
    return sum(1 for digit in str(abs(n)) if int(digit) % 2 == 0)
total = sum_odd_digit_3digits()
even_count = count_even_digits(total)

print("Сумма трёхзначных чисел с нечётными цифрами:", total)
print("Количество чётных цифр в этой сумме:", even_count)
```

```python
from math import factorial

def sum_even_factorials(m, n):
    total = 0
    for x in range(m, n + 1):
        if x % 2 == 0:
            total += factorial(x)
    return total

m = int(input("Введите m: "))
n = int(input("Введите n: "))
print(f"Сумма факториалов чётных чисел от {m} до {n} равна: {sum_even_factorials(m, n)}")
```
