```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Зачислено {amount:.2f}. Новый баланс: {self.balance:.2f}")
        else:
            print("Сумма пополнения должна быть положительной")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Выведено {amount:.2f}. Новый баланс: {self.balance:.2f}")
        elif amount > self.balance:
            print("Недостаточно средств.")
        else:
            print("Сумма снятия должна быть положительной.")

    def display_info(self):
        print(f"Номер банковского счета: {self.account_number}, Баланс: {self.balance:.2f}")

acc1 = BankAccount("123456577", 1000)
acc1.deposit(200)
acc1.withdraw(500)
acc1.withdraw(300)
acc1.withdraw(900)
acc1.display_info()
```

```python
class Document:
    def __init__(self, title, content):
        self.title = title
        self.content = content

    def display_info(self):
        print(f"Название: {self.title}")
        print(f"Содержание: {self.content}")


class Company(Document):
    def __init__(self, title, content, address):
        super().__init__(title, content)
        self.address = address

    def update_address(self, new_address):
        self.address = new_address
        print(f"Адрес обновлён: {self.address}")

    def display_info(self):
        super().display_info()
        print(f"Адрес: {self.address}")

company = Company("StailFX", "Продажа велосипедов", "ул. Ямашева, 10, Казань")
company.display_info()

company.update_address("Площадь Тукая, 25, Казань")
company.display_info()
```

```python
class Product:
    def __init__(self, name, factory):
        self.name = name
        self.factory = factory

    def show_info(self):
        print(f"Продукт: {self.name}")
        print(f"Произведено на фабрике: {self.factory.name}")


class Factory:
    def __init__(self, name):
        self.name = name

    def create_product(self, product_name):
        return Product(product_name, self)


factory = Factory("StailFX")
product1 = factory.create_product("СмартФон")
product2 = factory.create_product("Ноутбук")
product1.show_info()
product2.show_info()
```

```python
class Product:
    def __init__(self, name, serial_number):
        self.name = name
        self.serial_number = serial_number

    def show_info(self):
        print(f"Продукт: {self.name}")
        print(f"Серийный номер: {self.serial_number}")


class ProductionLine:
    def __init__(self, name):
        self.name = name

    def produce_product(self, product_name, serial_number):
        return Product(product_name, serial_number)


product1 = Product("Телефон", "A12345")
product2 = Product("Ноутбук", "B67890")
line = ProductionLine("Основная линия")
product3 = line.produce_product("Планшет", "C11223")
product1.show_info()
product2.show_info()
product3.show_info()
```

```python
from abc import ABC, abstractmethod

class Function(ABC):
    @abstractmethod
    def evaluate(self, x):
        pass


class LinearFunction(Function):
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def evaluate(self, x):
        return self.a * x + self.b


class QuadraticFunction(Function):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def evaluate(self, x):
        return self.a * x**2 + self.b * x + self.c


linear = LinearFunction(2, 3)
quadratic = QuadraticFunction(1, -2, 1)
x_value = 5
print(f"Линейная функция при x = {x_value}: {linear.evaluate(x_value)}")
print(f"Квадратичная функция при x = {x_value}: {quadratic.evaluate(x_value)}")
```

```python
from abc import ABC, abstractmethod

class IAccount(ABC):
    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def get_account_info(self):
        pass


class Account(IAccount):
    def __init__(self, account_number, balance=0.0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Внесено {amount:.2f}. Новый баланс: {self.balance:.2f}")
        else:
            print("Сумма для внесения должна быть положительной.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Снято {amount:.2f}. Новый баланс: {self.balance:.2f}")
        else:
            print("Недостаточно средств или некорректная сумма.")

    def get_account_info(self):
        print(f"Номер счёта: {self.account_number}")
        print(f"Баланс: {self.balance:.2f}")


account = Account("12345678", 1000.0)
account.get_account_info()
account.deposit(500)
account.withdraw(300)
account.get_account_info()
```

```python
class Company:
    def __init__(self, name):
        self.name = name

    def generate_report(self):
        print(f"Отчёт о деятельности компании {self.name}.")


class TechCompany(Company):
    def generate_report(self):
        print(f"Технический отчёт компании {self.name}: анализ данных.")


class ManufacturingCompany(Company):
    def generate_report(self):
        print(f"Производственный отчёт компании {self.name}: финансовое моделирование.")

tech_company = TechCompany("StailFX")
manufacturing_company = ManufacturingCompany("StailFX2")
tech_company.generate_report()
manufacturing_company.generate_report()
```
